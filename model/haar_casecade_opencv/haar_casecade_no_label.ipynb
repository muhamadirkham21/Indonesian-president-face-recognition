```python
import cv2 as cv
```

```python
# load image and rescale it to (500, 500) dimensions
img = cv.imread('/home/irkham/PROJECT/Indonesian-president-face-recognition/data/jokowi/jokowi2.jpg')
rescale = cv.resize(img, (500, 500), interpolation=cv.INTER_CUBIC)

# convert to greyscale image
gray = cv.cvtColor(rescale, cv.COLOR_BGR2GRAY)
haar_casecade = cv.CascadeClassifier('haarcascade_face.xml')

# predict and draw the bounding boxes
faces_rect = haar_casecade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=7)

for (x, y, w, h) in faces_rect:
    cv.rectangle(rescale, (x, y), (x+w, y+h), (0, 255, 0), thickness=2)

# show the actual and result image
cv.imshow('image', rescale)
cv.imshow('result', rescale)

cv.waitKey(0)
cv.destroyAllWindows()
```
